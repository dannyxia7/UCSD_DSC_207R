### 4. Subplots

Subplots allow us to have several graphs in one view which may help us analyse quickly and efficiently. Our goal is to create 2 subplots to understand trends by neighbourhood in San Diego. Let us start by setting up our subplots.



```python
#Run this cell
fig, axes = plt.subplots(1, 2, figsize=(20, 5))
```

#### Bar Graph
For the first of 2 subplots, Plot top 10 neighbourhoods in San Diego with most number of AirBnB units available. Hence this would be a bar graph with neighbourhood and counts. This may require 2 steps -

####  4.1. Extracting the df for `top_10_neighbourhoods`. (Recall value_counts() and slicing from Week 3) [0.5pt]

```python
top_10_neighbourhoods = listings_df.value_counts('neighbourhood')[:10]
```

<!-- BEGIN QUESTION -->


 #### 4.2. Plot the graph [1pt]
  1.   Provide axis labels
  2.   Provide a chart title/caption
  3.   If done correctly, we should see a graph with a non-increasing count of airbnb by neighbourhoods.

image.png

```python
# WRITE YOUR CODE HERE
sns.barplot(data=top_10_neighbourhoods)
plt.xticks(rotation=90)
plt.xlabel('Neighbourhoods')
plt.ylabel('Number of Listings')
plt.title('Number of Listings by Neighbourhoods')
```

<!-- END QUESTION -->

#### Boxplot

For the second subplot, we want to find a distribution of prices of airbnbs in the `top_10_neighbourhoods`. To perform this you may consider 2 steps -

####  4.3. Creating a `filtered_df` that has all the data for only the `top_10_neighbourhoods`. Hint - the isin() method may be useful here. [0.5pt]

```python
filtered_df = listings_df[listings_df['neighbourhood'].isin(list(top_10_neighbourhoods.index))]
```

```python
grader.check("Q4.3")
```

<!-- BEGIN QUESTION -->


####  4.4. Plot boxplot with x-axis as neighbourhood and y-axis as price. [1pt]
  1.   Provide axis labels
  2.   Provide a chart title/caption
  3.   Set parameters `showfliers = False` to get a cleaner graph by removing the outliers.

image.png

```python
# WRITE YOUR CODE HERE
sns.boxplot(data=filtered_df, x='neighbourhood', y='price', showfliers=False, palette="husl")
plt.xticks(rotation=90)
plt.xlabel('Neighbourhood')
plt.ylabel('Price')
plt.title('Price of Listings by Neighbourhood')
```

#### 4.5 **Print the combined subplot.** The report should contain both the graphs created in part 4 as subplots (as shown below)
1. Use the code for bar chart created previously

3. Use the code for box plot created previously.

```python
# WRITE YOUR CODE HERE
fig, axes = plt.subplots(1, 2, figsize=(20, 5))
sns.barplot(data=top_10_neighbourhoods, ax=axes[0])
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)
axes[0].set_xlabel('Neighbourhoods')
axes[0].set_ylabel('Number of Listings')
axes[0].set_title('Number of Listings by Neighbourhoods')

sns.boxplot(data=filtered_df, x='neighbourhood', y='price', showfliers=False, palette="husl", ax=axes[1])
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90)
axes[1].set_xlabel('Neighbourhood')
axes[1].set_ylabel('Price')
axes[1].set_title('Price of Listings by Neighbourhood')

plt.tight_layout()
plt.show()
```
